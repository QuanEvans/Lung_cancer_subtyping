```python
import GEOparse
import sklearn.cluster
import numpy as np
import random
import pandas as pd
```

```python
#This code loads in the data
gse = GEOparse.get_GEO(filepath = "GSE10245_family.soft.gz")
```

```python
#This code transposes the gene IDs to make them the rows for analysis
matrix_data = gse.pivot_samples("VALUE").T
```

```python
OMP_NUM_THREADS=1
```

```python
#This code performs the k-clustering, splitting up the data into 2 clusters. The random state is to keep results consistent with Quancheng's code
results = sklearn.cluster.k_means(matrix_data, 2, random_state = 575)
```

```python
#The first part of the accuracy data frame is the standard labels, which this extracts
sample_labels = matrix_data.index
```

```python
#The second part of the accuracy data frame is the generated k-means cluster assignments
cluster_numbers = results[1]
```

```python
#This loop extracts metadata information on the diseases each of the genes was originally assigned to
charac_list = []
for gsm in gse.gsms.values():
    charac_list.append(gsm.metadata["characteristics_ch1"])
```

```python
#This loop does some cleanup on the disease assignments so that they can be used in the accuracy data frame
format_charac_list = []
for entry in charac_list:
    entry = entry[0]
    entry = entry.replace("disease state: ", "")
    format_charac_list.append(entry)
```

```python
#This code creates the accuracy data frame described in part 3 
label_df = pd.DataFrame([sample_labels, cluster_numbers, format_charac_list]).T
label_df.columns = ["Labels", "Cluster", "Subtype"]
```
