```python
from gse_tools.GSEs import GSEs
```

```python
filepath="./../data/GSE10245_family.soft.gz"
gse = GSEs(file_path=filepath)
```

```python
gse.set_seed(575)
gse.train_model().predict(testOnTrain=True).accuracy#(i) train the model on all the data and compute the accuracy of the model on that data
```

```python
gse.train_model(n_clusters=2,train_frac=0.5).predict(testOnTrain=True).accuracy #(ii) train the model on half of the data and compute the accuracy of the model on that data
```

```python
gse.train_model(n_clusters=2, train_frac=0.5).predict().accuracy #(iii) compute the model accuracy built at point (ii) on the other half of the data (test data)
```
